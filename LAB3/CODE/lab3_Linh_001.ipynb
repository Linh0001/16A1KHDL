Đọc file stocks2.csv vào DataFrame stocks2

```python
stocks2=pd.read_csv('D:/16A1HN/LAB3/DATA/stocks2.csv')
stocks2
```

Gộp stocks1 và stocks2 vào dataframe stocks

```python
stocks=pd.concat([stocks1,stocks2])
stocks
```

Tính giá trung bình(open,high,low,close) cho mỗi ngày

```python
stocks_group=stocks.groupby('date')
kq=stocks_group.aggregate({'open':'mean','high': 'mean','low': 'mean','close': 'mean'})
kq
```

LAB 3.4

Đọc file companies.csv và DataFrame companies

```python
companies=pd.read_csv('D:/16A1HN/LAB3/DATA/companies.csv')

#Hiển thị 5 dòng đầu
print('Năm dòng đầu:\n',companies.iloc[:5])
```

Kết hợp stocks và companies dựa trên cột chung là symbol

```python
gp=pd.merge(stocks,companies,left_on='symbol',right_on='name').drop('name',axis=1)
gp
```

Tính giá đóng cửa(close) trung bình cho mỗi công ty

```python
#Nhóm theo tên công ty
symbol_group=gp.groupby('symbol')
#Tính trung bình bằng phương thức aggregate
mean_close=symbol_group.aggregate({'close':'mean'})
print('Hiển thị kết quả 5 dòng đầu:\n')
mean_close.iloc[:5]
#cáchkhac
#stocks.groupby(['symbol'])['close'].mean()
```

```python
pivot_table=stocks.pivot_table(index='date',columns='symbol',values='close',aggfunc='mean')
pivot_table
```

Thêm cột tính tổng volume giao dịch cho mỗi mã chứng khoán(symbol)

```python
pivot_stocks= stocks.pivot_table('close',index = 'date',columns='symbol',aggfunc='sum',margins=True,margins_name='total_volume')
pivot_stocks
```

```python
sorted_pivot_table = pivot_stocks.sort_values(by='total_volume', ascending=False)
sorted_pivot_table.head()
```
